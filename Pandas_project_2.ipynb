**Find the most powerful pokemon from the first 3 generations, of type water ?**

**What's the most powerful Dragon from the last two generations?**

```python
df.loc[
    ((df['Type 1'] == 'Dragon' ) | (df['Type 2'] == 'Dragon'))
     &
    (df['Generation'].isin([5,6]))


].sort_values( by = ['Total'] , ascending = False)
```

**Perform a selection in your Dataframe of all the Legendary pokemons that are of Type1 Fire. But select only the columns Name, Attack and Generation. Store the results in a variable legendary_fire_df.**
